```python
#Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.
import scipy.stats as stats

mean = 50
std_dev = 5
sample_size = 100  # Assuming a sample size of 100 for demonstration purposes

confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the margin of error
margin_of_error = stats.norm.ppf(1 - alpha / 2) * (std_dev / (sample_size ** 0.5))

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
#Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.
import numpy as np
from scipy.stats import norm
#population
n=500
#No.of indiusulas who smoke
smokers=60
#Portion of smokers in the sample 
p_hat=smokers/n

SE=np.sqrt((p_hat * (1 - p_hat)) / n)

Z_score=norm.ppf(0.975)

# Calculate the margin of error
margin_of_error=SE*Z_score
# Calculate the confidence interval
lower_boud=p_hat-margin_of_error
upper_boud=p_hat+margin_of_error
# Print the confidence interval
print("95% Confidence Interval for the true proportion of smokers:")
print(f"({lower_bound:.4f}, {upper_bound:.4f})")
```

```python

import numpy as np
from scipy.stats import norm

# Sample mean
sample_mean = 75

# Standard deviation
std_dev = 12

# Sample size (assuming a large enough sample for the Z-distribution)
sample_size = 100

# Confidence level
confidence_level = 0.90

# Calculate the standard error of the mean
standard_error = std_dev / np.sqrt(sample_size)

# Calculate the margin of error
margin_of_error = norm.ppf((1 + confidence_level) / 2) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("90% Confidence Interval:", confidence_interval)
```

```python
#Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Define range of values for chi-square distribution
x = np.linspace(0, 30)

# Calculate chi-square probability density function (PDF) for the given degrees of freedom
pdf = chi2.pdf(x, df)

# Plot chi-square distribution
plt.plot(x, pdf, label=f'Chi-square Distribution with {df} DF')

# Shade the area corresponding to a chi-square statistic of 15
x_shaded = np.linspace(0, 15)
plt.fill_between(x_shaded, chi2.pdf(x_shaded, df), color='orange', alpha=0.5, label='Chi-square Statistic of 15')

# Label axes
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density Function')

# Add a legend
plt.legend()

# Show plot
plt.show()
```

```python
#Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

import numpy as np
from scipy.stats import norm

# Sample size
n = 1000

# Number of people who prefer Coke in the sample
preferred_coke = 520

# Proportion of people who prefer Coke in the sample
p_hat = preferred_coke / n

# Calculate the standard error (SE) of the proportion
SE = np.sqrt((p_hat * (1 - p_hat)) / n)

# Z-score for 99% confidence level (two-tailed)
z_score = norm.ppf(0.995)

# Calculate the margin of error
margin_of_error = z_score * SE

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the confidence interval
print("99% Confidence Interval for the true proportion of people who prefer Coke:")
print(f"({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
#Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.
To conduct a chi-square goodness of fit test, you need to calculate the test statistic and compare it to the critical value at a given significance level. The test statistic is calculated as follows:

χ2=∑Ei​(Oi​−Ei​)2​

where Oi is the observed frequency in the i-th category and Ei is the expected frequency in the i-th category. The degrees of freedom for the test is the number of categories minus 1.

In this case, the researcher observed 45 tails in 100 flips, which is 45 tails out of a total of 100 flips. The expected frequency of tails is 50 (since there are 100 flips and each flip has a 50% chance of being tails). The test statistic is:

χ2=50(45−50)2​+50(55−50)2​=2.4

With 1 degree of freedom, the critical value at a 0.05 significance level is 3.84. Since the test statistic is less than the critical value, the researcher cannot reject the null hypothesis that the coin is fair. The observed frequencies match the expected frequencies of a fair coin.
```
