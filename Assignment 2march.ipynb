```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
```

```python
# A scatter plot is a two-dimensional data visualization that displays individual data points as dots, where each dot represents the values of two variables. It is useful for revealing patterns, correlations, or clusters in the data.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y)
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange and display multiple plots in a grid, specifying the number of rows and columns
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the first line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for the second line
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for the third line
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for the fourth line
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plot each line separately
plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')
plt.plot(x3, y3, label='Line 3')
plt.plot(x4, y4, label='Line 4')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')

 
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#A bar plot is a type of data visualization that represents categorical data with rectangular bars. Each bar's length corresponds to the value it represents, and the bars can be arranged either horizontally or vertically.
# A bar plot is used to visually compare and represent the values or frequencies of different categories using rectangular bars. It is effective for displaying categorical data and highlighting patterns, comparisons, or distributions.
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(y=company, width=profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')

plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
#A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution of a dataset. It provides a summary of the data's central tendency, spread, and identifies outliers
#Box plots are used to:

#Identify the central tendency and spread of a dataset.
#Detect skewness or asymmetry in the data.
#Identify potential outliers.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box1 and Box2')

# Show the plot
plt.show()
```
