```python
#Q1. How can you create a Bokeh plot using Python code?
 
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure ,show

x=[1,2,4,7,5]
y=[4,5,5,7,2]

p=figure(x_axis_label='x',y_axis_label='y')

circle=p.circle(x,y,legend_label='Objects',fill_color='red' ,size=50)

show(p)
```

```python
#Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

#In Bokeh terminology, these geometrical shapes are called gylphs. Bokeh plots constructed using bokeh.plotting interface use a default set of tools and styles. However, it is possible to customize the styles using available plotting tools.

x=[1,2,3,4,5]
y=[3,5,5,7,8]

p=figure(title='Show the use of glyphs',x_axis_label='x',y_axis_label='y')

circle =p.circle(x,y,legend_label='Objects',fill_color='blue',fill_alpha=0.7,line_color='red',size=50)
glyph=circle.glyph
glyph.fill_color='black'
glyph.size=29

show(p)
```

```python
#Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

#customize appearence of the bokeh plot
#Importing for the theme of the plot
from bokeh.io import curdoc

x=[1,2,4,7,5]
y=[4,5,5,7,2]

curdoc().theme='dark_minimal'

p=figure(x_axis_label='x',y_axis_label='y',max_width=500, height=350)

circle=p.circle(x,y,legend_label='Objects',fill_color='red' ,size=50)

show(p)
```

```python
# customize axes

x=[1,2,4,7,5]
y=[4,5,5,7,2]

curdoc().theme='light_minimal'

p=figure(x_axis_label='x',y_axis_label='y',max_width=500, height=350) 

circle=p.circle(x,y,legend_label='Objects',fill_color='red' ,size=50)
#Change something about x axis
p.xaxis.axis_label='x axis'
p.xaxis.axis_line_width=3
p.xaxis.axis_line_color='green'

#Change something about y axis
p.yaxis.axis_label='y axis'
p.yaxis.major_label_text_color='blue'
p.yaxis.axis_line_color='purple'



show(p)
```

```python
# Customize legend
x=[1,2,4,7,5]
y=[4,5,5,7,2]


p=figure(x_axis_label='x',y_axis_label='y',max_width=500, height=350) 

circle=p.circle(x,y,legend_label='Objects',fill_color='red' ,size=50)

p.legend.title='Showing'
p.legend.location='top_left'

# change legend style and color
p.legend.label_text_font = "times"
p.legend.label_text_font_style = "italic"
p.legend.label_text_color = "navy"

# change border and background of legend
p.legend.border_line_color = "navy"
p.legend.border_line_alpha = 0.8
p.legend.background_fill_color = "navy"
p.legend.background_fill_alpha = 0.2

show(p)
```
